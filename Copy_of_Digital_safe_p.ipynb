## **Digital safe Project**

## 1 **Introduction** <a id='introduction'></a>

In the 1920s and 1930s, safes were based solely on mechanical mechanisms that required turning a physical lock or using a physical key to open the safe. These mechanisms were relatively simple but not very secure, especially as new technological tools began to emerge.

In the 1980s and 1990s, the use of digital locks for safes began to spread, based on new technologies. A digital lock operates using a secret code system (PIN) or a smart card instead of a physical key.

These mechanisms often include a small display or keypad that allows the user to enter the code and activate the locking mechanism.

**Combinational Logic** is a field that deals with the use of logic gates to create electronic circuits that perform computational operations. Logic gates are basic components in digital circuits that implement Boolean functions such as XOR, NOT, OR, AND, and others. Each logic gate receives a binary input ("0" or "1") and produces a binary output according to the function it implements.

By combining logic gates, it is possible to create more complex circuits such as arithmetic operations, logical function implementations, memory, and central processing units (CPUs).
Combinational logic forms the foundation of modern technologies and plays a key role in the design and development of processors, memory, and other digital components used in a wide variety of applications—from simple electrical devices, to personal computers, mobile devices, and industrial systems.

Logical memory components have become an essential part of digital circuits, used for storing and processing binary data. These components include memory cells capable of maintaining a logical state of '0' or '1'.

Logical memory components are used in a variety of applications such as temporary data storage, storing internal system states, and executing complex computations. In addition, they are integrated into more advanced circuits such as RAM and ROM, where fast and reliable access to data is required.

Their main role is to enable digital systems to operate efficiently and stably, while preserving critical information required for various operations. These components are used in a wide range of applications, from industrial control systems to everyday electronic devices.

**Component – CD14516** is used as a **binary counter** capable of counting both up and down.
Using this component and its operation, it is possible to implement, among other things, a frequency divider.
A binary counter is a device that counts electrical pulses according to clock signals. It is typically built from a series of flip-flops—components with memory capabilities.
There are two main types of counters: synchronous and asynchronous.
In a synchronous counter, all the flip-flops are triggered by the same clock signal, whereas in an asynchronous counter, multiple clocks or the output of previous components in the system influence the behavior.
In addition, any binary counter can also function as a frequency divider, meaning it divides the input clock frequency by a certain value.
A cascade of flip-flops divides the clock frequency by: 2ⁿ, but by using logic components or alternative connection methods, division by other frequency values can also be implemented.

**The 555 timer** is an electronic component that contains multiple transistors and amplifiers integrated with resistors. It can be used to generate clock signals with control over their frequency and duty cycle (D/C). It can function as an oscillator or as a time delay generator.
The component consists of comparators, a trigger, a flip-flop, a discharge transistor, and an output.
The 555 timer can be implemented in two different modes:
1.Monostable mode – The component generates a single pulse in response to a trigger signal applied to pin 3.
2.Astable mode – The component operates as a free-running oscillator, generating a continuous waveform without requiring external pulses. The frequency and duty cycle are determined by the values of the connected resistors and capacitors.

There are two main connection types:
Common Cathode – All the LED cathodes (negative terminals) are connected to a common ground.
Common Anode – All the LED anodes (positive terminals) are connected to a common power supply.

## 1.3 **System Requirements** <a id='Requirements'></a>


For proper operation of the system, several requirements must be met:

First, a delay mechanism is needed for both the first and second systems to prevent the safe from locking immediately.

Since the system does not operate with an external clock, an oscillator must be created using only electrical components and logic gates.

There is a need to amplify voltage and current for the second system. Therefore, it is necessary to understand and use a component that enables signal amplification, such as a transistor.

The final system must meet all requirements and should therefore be simple, efficient, and reliable, while keeping production costs low. Additionally, the components used should have a long lifespan to ensure the safe remains functional over the long term.
## 1.3.1 **System Operation Description**
**Step 1:**
When the correct code is entered, the safe opens.
A green light (LED) turns on and remains lit for 5 seconds.
**Step 2:**
When the reset button "*" is pressed, the mechanical lock opens for 5 seconds and the system is reset immediately.
The 7-segment display is activated to show the current status.
**Step 3:**
If an incorrect code is entered, the second light (yellow LED) turns on for 1 second to indicate the error.
**Step 4:**
If an incorrect code is entered three times, a buzzer sounds continuously for 3 seconds, and the number of errors is displayed.
A red LED bar turns on and remains lit for 1 second to indicate the failure.









## 2 **Safe Unlocking – Correct Code** <a id='Correct'></a>


Below are the hardware design stages for unlocking the safe:

Step 1:
The code is valid only for the correct combination. A light turns on for five seconds when the correct code is entered.

Step 2:
A countdown display simulates the safe locking after five seconds, and immediate locking also occurs when the “*” or “#” buttons are pressed.

The analysis is divided into several parts:

First, the keypad is represented as a switch box – the red box in the figure above.
The digits 0–9 on the keypad, from left to right, are connected to OR gates. As a result, pressing any of these digits produces a logical “1”. The output of the OR gates is fed into the Clock input of the four shift registers to create a manually triggered clock signal.

Additionally, the buttons for the code “304” are connected to the shift registers as stored data – that is, the correct code. These switches are connected to the data input pins of the shift registers as preloaded information.

Each button press advances the shift registers by 1. If the code “304” is entered in the correct order, the data is pushed out such that:

The Q0 output of the first shift register has a logical “1” and the pushed-out digit is “4”

The Q1 output of the second shift register has a logical “1” and the digit is “0”

The Q2 output of the third shift register has a logical “1” and the digit is “3”

Thus, a logical “1” is obtained at the output of the shift registers only for the correct combination.

The outputs of the shift registers are connected to an AND gate, which sends a command only when all three outputs match the correct combination and are “1”.

The AND gate is connected to a NOT gate, because the trigger of the 555 timer receives a NOT signal – i.e., it is activated by a low voltage.

The timer is configured in monostable mode. In this mode, it produces a single pulse, which is needed for the system. A pulse duration of five seconds is required.

The generated pulse duration is calculated using the formula:
T = 1.1 × R × C (1)
Where:

T = time duration

R = resistance

C = capacitance

After the five-second pulse is generated, the green LED lights up for five seconds.

The Astable timer is connected to one input of an AND gate, while the other input of the AND gate is connected to the output of the Monostable timer.
The output of the AND gate is connected to the clock input of the binary counter, so that the counter’s clock signal is activated only when both the Monostable and the Astable timers are active.

In parallel, two buttons were added to simulate the "*" and "#" keys.
These buttons are connected to an OR gate and a diode, which are then connected to the RESET pin of the Monostable timer and also to the RESET pin of the binary counter, in order to reset the count.

In addition, the mentioned keys are also connected to an input of a second OR gate.
This second OR gate receives its other input from the output of the binary counter, and its output is connected to the p.e (preset enable) input pin.

The binary counter is configured to count down, as required for the second stage.
Its outputs are connected to a decoder to display the countdown.

The decoder is essential because the display component cannot interpret discrete voltages ("0" or "1" logic levels) and instead recognizes only binary words.
The decoder converts the electrical signals from the counter into a binary word that the display can understand.

At the end of this stage, we have a system that performs a countdown and opens the safe when the correct code is entered.

## 3 **Mistake Code system**  <a id='Mistake'></a>

Now, a system was designed to monitor incorrect code entries, in contrast to the previous system which monitors correct code entries.
The system was divided into two operational stages:

Stage 1 – The system detects an incorrect code and indicates it using a yellow LED that lights up for one second for each incorrect attempt.

Stage 2 – The system detects three consecutive incorrect code entries, displays a count of the incorrect key presses, activates a buzzer for three seconds, and turns on a red LED for one second.

The analysis was divided into several parts:

First, the third system consists of an AND gate with two inputs:

The first input is connected to the RESET of the 3-attempt detection circuit.

The second input is connected to the signal indicating an incorrect code entry, i.e., the output of a NOT gate that receives the signal from the 555 timer used for detecting correct codes.

The output of this AND gate is connected to a NOT gate and to the third 555 timer in the system, which operates in monostable mode to produce a one-second pulse for each incorrect code.
At the output of this stage, a yellow LED is connected to complete the implementation of Stage 3.

The fourth system, which completes the implementation of the project, also consists of several components:

A second 4015BD Shift Register is used as follows:

The output of the third 555 timer (from Stage 3) is connected to the clock input of the shift register.

The data input is connected to a logical “1.”

The reset pin is connected to the third output of the current shift register.

The third output of the shift register is also connected to a NOT gate, whose output is connected to the trigger pin of a new 555 timer, which generates a three-second pulse.

The first three outputs of the shift register are displayed using a LED BAR, also configured in a Common Cathode configuration.

The output of the new 555 timer is connected to two places:

A buzzer, which produces a sound for three seconds.

One input of another AND gate, whose second input is the output of the third system’s 555 timer.

At the output of this AND gate, a red LED is connected, which indicates that three incorrect code entries were made by lighting up for one second.

At the end of this stage, the entire system was successfully implemented in simulation.

## 7 **Issues in the Physical Implementation and Their Solutions**
 <a id='Issues'></a>

In the physical implementation, unlike the simulation which worked smoothly, several malfunctions occurred that required a deeper understanding of digital circuits. It is important to note that these issues were not related to the simulation but were actual, physical problems.

The first issue was button debounce:

In other words, a single press could be interpreted as multiple presses, which prevented the designed system from functioning correctly.

The debounce problem can be seen in the following figure:

צילום מסך 2025-07-23 ב-14.22.53.png

We solved the issue by using a relatively small capacitor that filters out unwanted noise and stabilizes the system. As a result, the button now outputs only a logical "0" or "1".

However, since the problem was not fully resolved, we implemented a monostable multivibrator to introduce a minimal delay for each press. This effectively eliminated the debounce issue.
As you can see in the following image:

צילום מסך 2025-07-23 ב-14.25.54.png

The second issue was that the output voltage for the correct code connected to the lock was only 3.5V instead of 5V, which prevented the lock from opening.

We resolved this by adding a BS170 MOSFET transistor to amplify the voltage. The drain pin was connected to a 5V VDD source, the source pin to ground, and the gate pin to the output of the 555 timer associated with the correct code.

Since this created an inverter, we added a NOT gate, ensuring the proper operation of the lock.

## 8 **Project Conclusions**

The difference between the measured and calculated values is partly due to the non-ideal behavior of real components, as opposed to theoretical calculations.

There is a significant deviation between the calculated values and those obtained from simulations in Multisim, most likely due to a built-in error in the software. Therefore, in circuits containing the 555 timer, the time constant is very different from the values used in practice.

Power amplification circuits are required; without them, the lock will not operate.

When many logic gates are chained together, a significant voltage drop occurs, which may cause a logical "1" to be interpreted as a "0" at the end of the process—contrary to the system's intended behavior. To identify or avoid this issue, either voltage amplifiers must be used or the design should be adjusted to prevent it altogether.

To solve complex design problems, one must view the entire system and its requirements as a whole. When each part is treated independently, it may lead to duplicated components and logic, resulting in an inefficient and cumbersome design that is very difficult to debug if faults occur.

To ensure the reliability of the final circuit, special attention must be paid to the verification of components before assembly. Additionally, emphasis should be placed on their placement within the circuit to make fault detection easier, if needed.

## 9 **Summary**  <a id='Summary'></a>
In this project, an advanced motherboard for an electronic safe was designed and developed using non-programmable electrical components only.

The project began with an in-depth analysis of system requirements, including the creation of a detailed block diagram describing the general structure of the motherboard and the interconnections between its components.

The design was then implemented using MULTISIM software, including detailed simulations of the required operations, feasibility tests, and implementation of logic using advanced electronic components.

The final system includes four main programs:

Correct code

Lock opening and closing

Incorrect code

Three incorrect code entries with buzzer alert

Compliance with safety standards and manufacturer specifications played a key role in the design. The chosen components were selected for their longer life expectancy, ensuring the final system is reliable over time.

Ultimately, component reliability is a core principle in electrical engineering. The final product was implemented with a focus on long-term stability and robustness, even if it involved using slightly more components compared to a more minimal (but less reliable) system.

The project was also constrained by a maximum of 15 components, a limitation that was fully respected. This constraint encouraged efficient design, allowing for the development of a more durable and practical system with a good cost-benefit ratio.

Throughout the project, several technical challenges were encountered, including:

Delay timings

Switching delays

Race conditions between components

Button bouncing

Limited lab time

Each of these challenges was fully addressed during the project.

Beyond technical issues, there were also discrepancies between simulation and physical implementation, requiring creative engineering thinking. This provides an important insight for future projects: how to better align simulation timings with real-world behaviors.

The final outcome demonstrated an efficient system that meets all technical and safety requirements, providing the user with a safe and intuitive experience.
This project serves as a successful example of the integration of precise engineering design, advanced technology, and hands-on problem-solving.
